```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv',encoding = 'unicode_escape')
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
# inplace=True (is use for saving the current done changes)
# drop null values
df.dropna(inplace=True)
```

```python
df.shape
```

```python
# change datatype of Amount
df['Amount']=df['Amount'].astype('int')
```

```python
# check datatype of Amount
df['Amount'].dtype
```

```python
df.columns
```

```python
# rename column
df.rename(columns={'Marital_Status':'shadi'})
```

```python
df.describe()
```

```python
# write describe for specific columns
df[['Age','Orders','Amount']].describe()
```

```python
df.columns
```

```python
# to plot graph
sns.countplot(x = 'Gender', data = df) 
```

```python
ax = sns.countplot(x='Gender',data=df)
# to show values on graph we have to use following method
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# if we want gender wise amount in descending order
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
# plot graph of above table
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

```python
# from above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than mens
```

```python
df.columns
```

```python
# now we can analyse data by age
# we can defferenciate male and female by using hue
sns.countplot(data=df,x='Age Group',hue='Gender')
```

```python
ax=sns.countplot(data=df,x='Age Group',hue='Gender')
# to show values on graph
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

```python
# from above graphs we can see that most of the buyers areof age group btween 26-35yrs female
```

```python
df.columns
```

```python
# total number of orders from top 10 states
sale_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
# we have to give graph size due to long states names
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Orders',data=sale_state)
```

```python
# total amount/sales from top 10 states
sale_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Amount',data=sale_state)
```

```python
# from above graphs we can see that most of the orders and total sale/amount are from uttar pradesh, maharashtra and karnataka respectively
```

```python
# analysis of married vs non married ppl sales
ax=sns.countplot(data=df,x='Marital_Status')
sns.set(rc={'figure.figsize':(3,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# analysis of amount with marital_status
sale_ms=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(data=sale_ms,x='Marital_Status',y='Amount',hue='Gender')
```

```python
# as above graphs we can see that most of the buyers are unmarried womens and they have high purchasing power
```

```python
df.columns
```

```python
# analysis of sale on the basis of occupation
ax=sns.countplot(data=df,x='Occupation')
sns.set(rc={'figure.figsize':(20,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sale_occupation=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sale_occupation,x='Occupation',y='Amount')
```

```python
# from above graphs we can see that most of the buyers are from IT,Healthcare and  Aviation sector
```

```python
# analysis of sale on the basis of product category
ax=sns.countplot(data=df,x='Product_Category')
sns.set(rc={'figure.figsize':(20,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sale_pc=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(data=sale_pc,x='Product_Category',y='Amount')
```

```python
# from above graphs we can see that most of the sold products are from food, clothing and electronics category
```

```python
# analysis of sale on the basis of product_id
sale_product=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.barplot(data=sale_product,x='Product_ID',y='Orders')
sns.set(rc={'figure.figsize':(20,5)})
```

```python
# conclusion
# as all over the graphs and analysis we can see that most buyers are the unmarried women,with age group 26-30, works in IT,Healthcare and Aviation,
# belongs to UP, Maharashtra and Karnataka who buys more food, clothing and electronics.
```
